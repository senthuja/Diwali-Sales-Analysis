```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# To render inline visualizations in Jupyter notebooks
%matplotlib inline
```

```python
# Load the CSV file containing Diwali Sales Data
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

# View the shape and first few rows of the dataset
df.shape
df.head(10)
```

```python
# Get a concise summary of the DataFrame
df.info()
```

```python
# Drop unrelated or unnecessary columns
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
df
```

```python
# Check for null values in the dataset
df.isnull().sum()
df
```

```python
# Drop rows with any null values
df.dropna(inplace=True)
df
```

```python
# Convert the 'Amount' column to integer type
df['Amount'] = df['Amount'].astype(int)
df
```

```python
# Standardize column names: convert to lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(" ", "_", regex=True)
df.columns
```

```python
# Rename specific columns for better clarity
df.rename(columns={'marital_status': 'civil_status'}, inplace=True)

# View descriptive statistics for the numerical columns
df.describe()
```

```python
# Describe specific columns like 'age', 'orders', and 'amount'
df[['age', 'orders', 'amount']].describe()
```

```python
#Gender Distribution
```

```python
# Plot a count of the gender of buyers
ax = sns.countplot(x='gender', data=df)
ax.set_title("Gender Distribution")

# Display bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plot the total amount spent by each gender
sales_gen = df.groupby('gender')['amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(x='gender', y='amount', data=sales_gen)
plt.title("Total Amount Spent by Gender")
```

```python
#Age Group Distribution
```

```python
# Plot a count of buyers by age group, separated by gender
ax = sns.countplot(x='age_group', hue='gender', data=df)
ax.set_title("Age Group Distribution by Gender")

# Display bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python

# Plot total amount spent by age group
sales_age = df.groupby('age_group')['amount'].sum().sort_values(ascending=False).reset_index()

sns.barplot(x='age_group', y='amount', data=sales_age)
plt.title("Total Amount Spent by Age Group")
```

```python
#Orders and Sales by State
```

```python
# Plot total number of orders from top 10 states
sales_state_orders = df.groupby('state')['orders'].sum().sort_values(ascending=False).head(10).reset_index()

sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(x='state', y='orders', data=sales_state_orders)
plt.title("Top 10 States by Number of Orders")
```

```python
# Plot total sales amount from top 10 states
sales_state_amount = df.groupby('state')['amount'].sum().sort_values(ascending=False).head(10).reset_index()

sns.set(rc={'figure.figsize': (15, 5)})
sns.barplot(x='state', y='amount', data=sales_state_amount)
plt.title("Top 10 States by Total Sales Amount")
```

```python
#Marital Status and Purchases
```

```python
# Plot count of buyers by marital status
ax = sns.countplot(x='civil_status', data=df)
sns.set(rc={'figure.figsize': (7, 5)})
ax.set_title("Distribution by Civil Status")

# Display bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plot total amount spent by marital status and gender
sales_civil_status = df.groupby(['civil_status', 'gender'])['amount'].sum().sort_values(ascending=False).reset_index()

sns.set(rc={'figure.figsize': (6, 5)})
sns.barplot(x='civil_status', y='amount', hue='gender', data=sales_civil_status)
plt.title("Total Amount Spent by Civil Status and Gender")
```

```python
#Occupation Distribution
```

```python
# Plot count of buyers by occupation
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(x='occupation', data=df)
ax.set_title("Distribution by Occupation")

# Display bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plot total amount spent by occupation
sales_occupation = df.groupby('occupation')['amount'].sum().sort_values(ascending=False).reset_index()

sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(x='occupation', y='amount', data=sales_occupation)
plt.title("Total Amount Spent by Occupation")
```

```python
#Product Category Analysis
```

```python
# Plot count of orders by product category
sns.set(rc={'figure.figsize': (20, 5)})
ax = sns.countplot(x='product_category', data=df)
ax.set_title("Orders by Product Category")

# Display bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Plot total amount spent by product category (Top 10)
sales_product_category = df.groupby('product_category')['amount'].sum().sort_values(ascending=False).head(10).reset_index()

sns.set(rc={'figure.figsize': (20, 5)})
sns.barplot(x='product_category', y='amount', data=sales_product_category)
plt.title("Top 10 Product Categories by Total Amount Spent")

```

```python
# Top 10 most sold products
fig, ax = plt.subplots(figsize=(12, 7))
df.groupby('product_id')['orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar', ax=ax)
ax.set_title("Top 10 Most Sold Products")
plt.show()
```
